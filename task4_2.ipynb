```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
import scipy.stats as sts
from scipy.stats import norm
```

```python
def na(n, p, amount):
    for _ in range(amount):
        m = 0
        for i in range(n):
            if random.random() < p:
                m += 1
        yield m/n
```

```python
def M(nums):
    return sum(nums)/len(nums)
```

```python
def D(nums):
    mean = M(nums)
    return (sum([i**2 for i in nums]) / len(nums) - mean**2)*len(nums)/(len(nums)-1)
```

```python
def hist(nums, a, b, intervals, width):
    bins = np.linspace(a, b, intervals+1)
    counts, bins = np.histogram(nums, bins)
    return list([i/len(nums)/width for i in counts]), bins
```

```python
def f(x, mat, s):
    return (1 / (s * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mat) / s) ** 2)
```

```python
amount_array = [10, 20, 50, 100, 200, 500, 1000]
n_array = [50, 100, 200]
p=0.5
for n in n_array:
    for amount in amount_array:
        nums = []
        for t in na(n, p, amount):
            nums.append(t)
        mid = M(nums)
        var = 4 * D(nums) ** 0.5
        #print(f": .={M(nums)} ={D2(nums)},  ={D2(nums)**0.5}")
        
        fig, ax = plt.subplots()
        frequinces, bins = hist(nums, mid-var, mid + var, 10, 1)
        #print(sum([i * (2*var/10) for i in frequinces]))
        
        plt.bar(bins[:-1], frequinces, align='edge', width=2 * var / 10, color='c')
        
        x = np.linspace(mid - var, mid + var, 100)
        plt.plot(x, [f(i, M(nums), D(nums) ** 0.5)*(2*var/ 10) for i in x], 'r')
        
        #ax.set_xticks([float("{:.2f}".format(bin)) for bin in bins])
        
plt.show()
```
