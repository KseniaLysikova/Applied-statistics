```python
import random
import math
import matplotlib.pyplot as plt
```

```python
ds = []
pis = []
num_tests = 5

for d in range(3, 7):
    ds.append(10**d)
    cur_pi = []
    for _ in range(num_tests):
        m = 0
        x0 = 1
        y0 = 2
        r0 = 5
        x_max = x0+r0
        x_min = x0-r0
        y_max = y0 + r0
        y_min = y0 - r0
        for i in range(10**d):
            x = (x_max - x_min) * random.random() + x_min
            y = (y_max - y_min) * random.random() + y_min

            if (x - x0)**2 + (y - y0)**2 - r0**2 < 10**(-6):
                m += 1
        cur_pi.append(4*m/10**d)
    pis.append(sum(cur_pi)/(len(cur_pi)))
```

```python
print(pis)
```

```python
delta = []
for pi in pis:
    tmp = abs(((pi - math.pi)/num_tests) * 100)
    delta.append(tmp)
```

```python
print(delta)
```

```python
plt.plot(ds, delta)
plt.show()
```
